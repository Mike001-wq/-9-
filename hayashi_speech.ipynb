```python
import requests
from bs4 import BeautifulSoup
import pickle
import os
import json
deepl_auth_key = os.environ.get("DeepL_Auth_Key")
```

```python
months = ['2111', '2112', '2201', '2202', '2203', '2204', '2205']
# "https://www.mofa.go.jp/mofaj/press/kaiken/gaisho/g_{month}.html"
```

```python
speeches = []
for month in months:
    url = f"https://www.mofa.go.jp/mofaj/press/kaiken/gaisho/g_{month}.html"
    response = requests.get(url)
    response.encoding = "UTF-8"
    soup = BeautifulSoup(response.text, "html.parser")
    # print(soup.prettify())
    speech_page_list = soup.find(id="pressconf").find_all('dl')
    for speech_page in speech_page_list:
        # print(speech_page)
        title = speech_page.find(class_="list-title").get_text().strip().split('）')[0]+'）'
        url = "https://www.mofa.go.jp" + speech_page.find(href=True)['href']
        # print(url)
        response = requests.get(url)
        response.encoding = "UTF-8"
        soup = BeautifulSoup(response.text, "html.parser")
        content = soup.find(id="maincontents")
        rightalign_other_language = content.find(class_="rightalign other-language")
        if rightalign_other_language:
            rightalign_other_language.clear()
        social_btn_top = content.find(class_="social-btn-top")
        social_btn_top.clear()
        content = content.get_text().replace('\r\n', '').replace('\t', '').strip('\n')
        # print(content)
        if "茂木" in title:
            continue
        speech = {
            "source": url,
            "title": title,
            "chinese_content": None,
            "japanese_content": content,
        }
        speeches.append(speech)
# print(speeches)
```

```python
def deepl_oauth(r):
    r.headers["Authorization"] = f"DeepL-Auth-Key {deepl_auth_key}"
    return r
def translator(text):
    # response = requests.get(url='https://api.deepl.com/v2/usage', auth=deepl_oauth)
    # print(response.text)
    data = {
        "text": text,
        "source_lang": "JA",
        "target_lang": "ZH"
    }
    response = requests.get(url='https://api.deepl.com/v2/translate', auth=deepl_oauth, data=data)
    result = json.loads(response.text)
    return result['translations'][0]['text']
# translator('test')
def translator_usage():
    response = requests.get(url='https://api.deepl.com/v2/usage', auth=deepl_oauth)
    print(response.text)
```

```python
translator_usage()
cnt = 0
for speech in speeches:
    cnt += 1
    print(f"{cnt}/{len(speeches)}")
    japanese_content = speech['japanese_content']
    chinese_content = translator(japanese_content)
    speech['chinese_content'] = chinese_content
print(speeches)
translator_usage()
```

```python
hayashi_speech_pk_path = f"hayashi_speech.pk"
hayashi_speech_pk = open(hayashi_speech_pk_path, 'wb')
pickle.dump(speeches, hayashi_speech_pk)
hayashi_speech_pk.close()
```
