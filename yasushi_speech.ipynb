```python
import requests
from bs4 import BeautifulSoup
import pickle
import os
import json
deepl_auth_key = os.environ.get("DeepL_Auth_Key")
```

```python
months = ['2110', '2111', '2112', '2201', '2202', '2203', '2204', '2205']
# "https://www.soumu.go.jp/menu_news/kaiken/{month}.html"
```

```python
speeches = []
for month in months:
    url = f"https://www.soumu.go.jp/menu_news/kaiken/{month}.html"
    response = requests.get(url)
    response.encoding = "Shift_JIS"
    soup = BeautifulSoup(response.text, "html.parser")
    # print(soup.prettify())
    # print(soup.find(class_="icolistA"))
    speech_page_list = soup.find(class_="idxLst").find_all('dd')
    for speech_page in speech_page_list:
        print(speech_page)
        url = "https://www.soumu.go.jp" + speech_page.find(href=True)['href']
        response = requests.get(url)
        print(url)
        response.encoding = "Shift_JIS"
        soup = BeautifulSoup(response.text, "html.parser")
        # print(soup.prettify())
        content = soup.find(class_="contentsBody")
        title = ""
        h1 = content.find("h1")
        h1.clear()
        h2 = content.find("h2")
        title += h2.get_text()
        h2.clear()
        date = content.find("p", class_="rightPosT")
        title += date.get_text()
        date.clear()
        content = content.get_text().replace('\r\n', '').replace('\t', '').strip('\n')
        # print(title)
        # print(content)
        if "武田" in title:
            continue
        speech = {
            "source": url,
            "title": title,
            "chinese_content": None,
            "japanese_content": content,
        }
        speeches.append(speech)
# print(speeches)
```

```python
def deepl_oauth(r):
    r.headers["Authorization"] = f"DeepL-Auth-Key {deepl_auth_key}"
    return r
def translator(text):
    # response = requests.get(url='https://api.deepl.com/v2/usage', auth=deepl_oauth)
    # print(response.text)
    data = {
        "text": text,
        "source_lang": "JA",
        "target_lang": "ZH"
    }
    response = requests.get(url='https://api.deepl.com/v2/translate', auth=deepl_oauth, data=data)
    result = json.loads(response.text)
    return result['translations'][0]['text']
# translator('test')
def translator_usage():
    response = requests.get(url='https://api.deepl.com/v2/usage', auth=deepl_oauth)
    print(response.text)
```

```python
translator_usage()
cnt = 0
for speech in speeches:
    cnt += 1
    print(f"{cnt}/{len(speeches)}")
    japanese_content = speech['japanese_content']
    chinese_content = translator(japanese_content)
    speech['chinese_content'] = chinese_content
print(speeches)
translator_usage()
```

```python
yasushi_speech_pk_path = f"yasushi_speech.pk"
yasushi_speech_pk = open(yasushi_speech_pk_path, 'wb')
pickle.dump(speeches, yasushi_speech_pk)
yasushi_speech_pk.close()
```
